## Computing similarity score

```python
# Calculate similarity scores
def compute_similarity_score(ratings, movie1, movie2):
    common_users = ratings[(ratings[movie1].notna()) & (ratings[movie2].notna())].index
    if len(common_users) == 0:
        return 0
    X = ratings[movie1][common_users]
    Y = ratings[movie2][common_users]
    distance = np.linalg.norm(X - Y)
    return 1 / (1 + distance)
```

## Initializing DataFrame with NaN values and filling with actual scores

```python
# Initialize DataFrame with NaN values
movie_similarity_scores = pd.DataFrame(index=movie_ratings.columns, columns=movie_ratings.columns)

# Calculating and filling in similarity scores
for movie1 in movie_ratings.columns:
    for movie2 in movie_ratings.columns:
        if movie1 != movie2:
            similarity_score = compute_similarity_score(movie_ratings, movie1, movie2)
            movie_similarity_scores.loc[movie1, movie2] = similarity_score

# Displaying the DataFrame
movie_similarity_scores
```

## Computing average simmilarity to display bar graph showing average similarity per movies

```python
# Calculate the average similarity per movie
avg_similarity_per_movie = movie_similarity_scores.mean(axis=1)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
avg_similarity_per_movie.plot(kind='bar', color='Orange', edgecolor='black')
plt.title('Average Similarity per Movie')
plt.xlabel('Movie Title')
plt.ylabel('Average Similarity Score')
plt.xticks(rotation=90)
plt.show()

```

## Creating function that recommend movies

```python
# Creating a function for movie recommendations based on item-based collaborative filtering
def recommend_movies(user_ratings, ratings, movie_similarity_scores, num_recommendations=5):
    recommendations = []

    for movie_title in ratings.columns:
        if pd.isna(user_ratings.get(movie_title)):
            predicted_rating = sum(user_ratings[i] * movie_similarity_scores[movie_title][i]
                                   for i in ratings.columns
                                   if not pd.isna(user_ratings.get(i)) and not pd.isna(movie_similarity_scores[movie_title][i]))
            denominator = sum(np.abs(movie_similarity_scores[movie_title][i])
                              for i in ratings.columns
                              if not pd.isna(user_ratings.get(i)) and not pd.isna(movie_similarity_scores[movie_title][i]))

            if denominator != 0:
                recommendations.append({'title': movie_title, 'predicted_rating': predicted_rating / denominator})

    top_recommendations = pd.DataFrame(recommendations).sort_values(by='predicted_rating', ascending=False).head(num_recommendations)
    return top_recommendations
```

## Providing user inputs

```python
# Privide user inputs to get recommendations
# user_input = {'Star Wars (1977)': 3}
# user_input = {'Restoration (1995)': 5}
# user_input = {'Toy Story (1995)': 4}
# user_input = {'Just Cause (1995)': 5}
# user_input = {'Men in Black (1997)': 3}
user_input = {'L.A. Confidential (1997)': 3}
top_recommendations = recommend_movies(user_input, movie_ratings, movie_similarity_scores, num_recommendations=5)
top_recommendations = top_recommendations.set_index('title')
print("You may like: \n")
print(top_recommendations)
```

## Calculating Mean Squared Error and Root Mean Squared Error

```python
from sklearn.metrics import mean_squared_error

# Trimming the dataset to include only the last 10000 rows
test_dataset = rating_dataset.tail(90000)

# Example: Replace 'predicted_ratings' with the actual variable containing your predictions
predicted_ratings = recommend_movies(user_input, movie_ratings, movie_similarity_scores, num_recommendations=5)
merged_datas = pd.merge(test_dataset, predicted_ratings, on='title')

# Calculating Mean Squared Error
mse = mean_squared_error(merged_datas['rating'], merged_datas['predicted_rating'])

# Calculating Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```
